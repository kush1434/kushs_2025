# Popcorn Hack 1

```python
public class Dog {
    // Original properties
    String name;
    int age;
    
    // New property for breed
    String breed;
    
    // Constructor
    public Dog(String name, int age, String breed) {
        this.name = name;
        this.age = age;
        this.breed = breed;
    }
    
    // Original method
    public void bark() {
        System.out.println(name + " says: Woof!");
    }
    
    // New method to print dog info with breed
    public void printBreedInfo() {
        System.out.println("My dog " + name + " is a " + breed + "!");
    }
}

// To use it:
class DogDemo {
    public static void main(String[] args) {
        System.out.println("=== Popcorn Hack #1 Demo ===");
        
        // Create a new Dog object
        Dog myDog = new Dog("Buddy", 3, "Golden Retriever");
        
        // Call the methods
        myDog.bark();
        myDog.printBreedInfo();
        
        // Create another dog to show it works
        Dog anotherDog = new Dog("Max", 5, "German Shepherd");
        anotherDog.printBreedInfo();
        
        System.out.println();
    }
}
```

# Popcorn Hack 2

```python
class Counter {
    private int count;
    
    // Original add method
    public void add(int x) {
        count += x;
    }
    
    // New method: subtraction
    public void subtract(int x) {
        count -= x;
    }
    
    // New method: multiplication
    public void multiply(int x) {
        count *= x;
    }
    
    // New method: division
    public void divide(int x) {
        if (x != 0) {
            count /= x;
        } else {
            System.out.println("Error: Cannot divide by zero!");
        }
    }
    
    // Original getter method
    public int getCount() {
        return count;
    }
    
    // Helper method to display current count
    public void displayCount() {
        System.out.println("Current count: " + count);
    }
}

// Use it:
class CounterDemo {
    public static void main(String[] args) {
        System.out.println("=== Popcorn Hack #2 Demo ===");
        
        Counter c = new Counter();
        
        // Method 1: Add
        c.add(10);
        System.out.print("After adding 10: ");
        c.displayCount();
        
        // Method 2: Subtract
        c.subtract(3);
        System.out.print("After subtracting 3: ");
        c.displayCount();
        
        // Method 3: Multiply
        c.multiply(4);
        System.out.print("After multiplying by 4: ");
        c.displayCount();
        
        // Method 4: Divide
        c.divide(2);
        System.out.print("After dividing by 2: ");
        c.displayCount();
        
        // Method 5: Get count (using original getter)
        int finalValue = c.getCount();
        System.out.println("Final value using getCount(): " + finalValue);
        
        // Test division by zero protection
        System.out.println("\nTesting division by zero:");
        c.divide(0);
        c.displayCount();
    }
}
```

# Homework Hack

```python
public class StudentGradeTracker {
    public static void main(String[] args) {
        System.out.println("=== Student Grade Tracker System ===\n");
        
        // Create first student: Emma Rodriguez
        System.out.println("Creating student: Emma Rodriguez");
        Student emma = new Student("Emma Rodriguez");
        System.out.println("Student created successfully!\n");
        
        // Add grades for Emma
        System.out.println("--- Adding Grades for Emma ---");
        emma.addGrade(95);
        emma.addGrade(88);
        emma.addGrade(92);
        emma.addGrade(85);
        System.out.println();
        
        // Print Emma's report
        System.out.println("--- Emma's Grade Report ---");
        emma.printReport();
        System.out.println("\n========================================\n");
        
        // Create second student: James Wilson
        System.out.println("Creating student: James Wilson");
        Student james = new Student("James Wilson");
        System.out.println("Student created successfully!\n");
        
        // Add grades for James
        System.out.println("--- Adding Grades for James ---");
        james.addGrade(78);
        james.addGrade(82);
        james.addGrade(75);
        System.out.println();
        
        // Print James's report
        System.out.println("--- James's Grade Report ---");
        james.printReport();
        System.out.println("\n========================================\n");
        
        // Final summary
        System.out.println("Final Summary:");
        System.out.printf("%s - Average: %.2f (%s)%n", 
            emma.name, emma.getAverage(), emma.getLetterGrade());
        System.out.printf("%s - Average: %.2f (%s)%n", 
            james.name, james.getAverage(), james.getLetterGrade());
    }
}

/**
 * Student class that tracks a student's grades and calculates statistics.
 * Each student object maintains their own grade data independently.
 */
class Student {
    // Instance variables - each Student object has its own copy
    String name;              // Student's full name
    int totalPoints;          // Sum of all grade points earned
    int numAssignments;       // Count of assignments completed
    
    /**
     * Constructor: Initializes a new Student with a name
     * Sets totalPoints and numAssignments to 0 initially
     * 
     * @param name The student's full name
     */
    public Student(String name) {
        this.name = name;
        this.totalPoints = 0;
        this.numAssignments = 0;
    }
    
    /**
     * Void instance method: Adds a grade to this student's record
     * Updates both totalPoints and numAssignments
     * 
     * @param points The points earned on an assignment
     */
    public void addGrade(int points) {
        totalPoints += points;
        numAssignments++;
        System.out.println("Adding grade: " + points + " points");
    }
    
    /**
     * Non-void instance method: Calculates current grade average
     * Returns 0.0 if no assignments have been completed (avoids division by zero)
     * 
     * @return The student's current average as a double
     */
    public double getAverage() {
        if (numAssignments == 0) {
            return 0.0;
        }
        return (double) totalPoints / numAssignments;
    }
    
    /**
     * Non-void instance method: Determines letter grade based on average
     * Uses standard grading scale: A(90+), B(80-89), C(70-79), D(60-69), F(<60)
     * 
     * @return A String representing the letter grade (A, B, C, D, or F)
     */
    public String getLetterGrade() {
        double avg = getAverage();
        
        if (avg >= 90) {
            return "A";
        } else if (avg >= 80) {
            return "B";
        } else if (avg >= 70) {
            return "C";
        } else if (avg >= 60) {
            return "D";
        } else {
            return "F";
        }
    }
    
    /**
     * Void instance method: Prints a formatted report of student's performance
     * Displays name, total points, assignments completed, average, letter grade,
     * and a status message based on performance
     */
    public void printReport() {
        double avg = getAverage();
        String letter = getLetterGrade();
        
        System.out.println("Student Name: " + name);
        System.out.println("Total Points: " + totalPoints);
        System.out.println("Assignments Completed: " + numAssignments);
        System.out.printf("Current Average: %.2f%n", avg);
        System.out.println("Letter Grade: " + letter);
        
        // Display status message based on letter grade
        System.out.print("Status: ");
        switch (letter) {
            case "A":
                System.out.println("Excellent work!");
                break;
            case "B":
                System.out.println("Great job!");
                break;
            case "C":
                System.out.println("Keep working hard!");
                break;
            case "D":
                System.out.println("Need to improve - seek help!");
                break;
            case "F":
                System.out.println("Urgent attention needed!");
                break;
        }
    }
}
```

## Code Explanation

### Instance Variables
The `Student` class contains three instance variables that store each student's individual data:
- `name`: Stores the student's full name
- `totalPoints`: Accumulates all grade points earned
- `numAssignments`: Tracks how many assignments have been completed

### Constructor
The constructor initializes a new `Student` object with a name and sets `totalPoints` and `numAssignments` to 0.

### Instance Methods

#### 1. `addGrade(int points)` - Void Method
- Adds a grade to the student's record
- Updates `totalPoints` by adding the new points
- Increments `numAssignments` counter
- Prints confirmation message

#### 2. `getAverage()` - Non-Void Method (returns double)
- Calculates the current grade average
- Includes division-by-zero protection
- Returns the average as a double

#### 3. `getLetterGrade()` - Non-Void Method (returns String)
- Determines letter grade based on average
- Uses standard grading scale (A: 90-100, B: 80-89, C: 70-79, D: 60-69, F: <60)
- Returns a single letter as a String

#### 4. `printReport()` - Void Method
- Displays a formatted report of the student's performance
- Shows name, total points, assignments completed, average, and letter grade
- Includes personalized status message based on performance
